```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from datetime import datetime
from datetime import timedelta
```

```python
# datetime function
def yday2date(year, month, day, hour, min, second):
    return pd.to_datetime(year + '-' + month + '-' + day + ' ' + hour + ':' + min + ':' + second,
                          format='%Y-%m-%d %H:%M:%S')
```

```python
file='/content/drive/MyDrive/ATMDP/Water_data/waterT_SMEAR_Kuivajarvi_120101_230630_30min.csv'
```

```python
file_data=pd.read_csv(file, sep=",",engine='python',parse_dates={'Date': ['Year', 'Month','Day','Hour','Minute','Second']},
                         date_parser=yday2date)
```

```python
dataframe1=pd.DataFrame(file_data)
dataframe1['Tdifference']=dataframe1['KVJ_META.Tw02']-dataframe1['KVJ_META.Tw120']
```

```python
import plotly.graph_objects as go
import pandas as pd
import plotly.io as pio

# Set the renderer for Google Colab
pio.renderers.default = 'colab'

# Filter out data for June 2018
dataframe_filtered = dataframe1[~((dataframe1['Date'] >= '2018-06-01') & (dataframe1['Date'] <= '2018-06-30'))]

# Create a Plotly figure
fig = go.Figure()

# Filter out data for June 2018
dataframe_filtered = dataframe1[~((dataframe1['Date'] >= '2018-06-01') & (dataframe1['Date'] <= '2018-06-30'))]

# Plot the filtered time series data for Tdifference
fig.add_trace(go.Scatter(
    x=dataframe_filtered['Date'],
    y=dataframe_filtered['Tdifference'],
    mode='lines',
    line=dict(color='blue', width=1),
    name='Tdifference'
))

# Plot additional depth temperature data
fig.add_trace(go.Scatter(
    x=dataframe1['Date'],
    y=dataframe1['KVJ_META.Tw02'],
    mode='lines',
    line=dict(color='grey', width=1),
    name='Depth 0.2m'
))
fig.add_trace(go.Scatter(
    x=dataframe1['Date'],
    y=dataframe1['KVJ_META.Tw30'],
    mode='lines',
    line=dict(color='cyan', width=1),
    name='Depth 3.0m'
))
fig.add_trace(go.Scatter(
    x=dataframe1['Date'],
    y=dataframe1['KVJ_META.Tw60'],
    mode='lines',
    line=dict(color='maroon', width=1),
    name='Depth 6.0m'
))
fig.add_trace(go.Scatter(
    x=dataframe1['Date'],
    y=dataframe1['KVJ_META.Tw120'],
    mode='lines',
    line=dict(color='purple', width=1),
    name='Depth 12.0m'
))

# Add vertical shaded spans (seasonal shading)
autumn_color = 'rgb(230, 159, 0)'
spring_color = 'rgb(0, 158, 115)'

# Define a function to add shaded areas without labels
def add_season_shading(start_date, end_date, color):
    fig.add_vrect(x0=start_date, x1=end_date, fillcolor=color, opacity=0.2, line_width=0)

# Adding the shaded areas for Autumn and Spring periods without labels
add_season_shading('2012-10-06', '2012-11-27', autumn_color)
add_season_shading('2013-04-27', '2013-05-07', spring_color)
add_season_shading('2013-10-12', '2013-11-24', autumn_color)
add_season_shading('2014-03-26', '2014-05-15', spring_color)
add_season_shading('2014-09-27', '2014-11-29', autumn_color)
add_season_shading('2015-04-20', '2015-05-17', spring_color)
add_season_shading('2015-10-03', '2015-11-27', autumn_color)
add_season_shading('2016-04-07', '2016-05-01', spring_color)
add_season_shading('2016-10-06', '2016-11-27', autumn_color)
add_season_shading('2017-03-29', '2017-05-18', spring_color)
add_season_shading('2017-10-03', '2017-12-13', autumn_color)
add_season_shading('2018-04-23', '2018-05-06', spring_color)
add_season_shading('2018-10-03', '2018-11-24', autumn_color)
add_season_shading('2019-04-23', '2019-05-11', spring_color)
add_season_shading('2021-04-20', '2021-05-11', spring_color)
add_season_shading('2021-09-24', '2021-11-22', autumn_color)
add_season_shading('2022-05-07', '2022-05-18', spring_color)
add_season_shading('2022-10-07', '2022-11-22', autumn_color)
add_season_shading('2023-04-21', '2023-05-06', spring_color)

# Add titles and labels
fig.update_layout(
    # title="Lake Kuivaj채rvi 2018 Turnover Periods",
    xaxis_title="Time",
    yaxis_title="Water temperature [째C]",
    font=dict(size=19),
    legend=dict(
        yanchor="bottom",  # Anchor legend to the bottom
        y=-0.6,            # Position it slightly below the plot
        xanchor="center",  # Center it horizontally
        x=0.6,             # Align to the center of the figure
        orientation="h"    # Horizontal orientation for the legend
    )
)

# Customize tick labels
fig.update_xaxes(tickangle=45, tickfont=dict(size=22))
fig.update_yaxes(tickfont=dict(size=22))

# Display the interactive Plotly figure
fig.show()


```
